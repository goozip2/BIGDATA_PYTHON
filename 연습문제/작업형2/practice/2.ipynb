```python
# 데이터 불러오기
import pandas as pd
X_train = pd.read_csv(r'C:\BigData\practice_data\연습문제\College_x_train.csv', encoding = 'cp949')
y_train = pd.read_csv(r'C:\BigData\practice_data\연습문제\College_y_train.csv', encoding = 'cp949')
X_test = pd.read_csv(r'C:\BigData\practice_data\연습문제\College_X_test.csv', encoding = 'cp949')
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
# id 빼두기
# 필요없는 칼럼 제거

result_id = X_test['ID']

X_train.drop(columns=['ID', 'Name'], inplace=True)
X_test.drop(columns=['ID', 'Name'], inplace=True)
y_train.drop(columns = ['ID'], inplace=True)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
```

```python
# 결측치 처리
col = X_train.columns
for i in col:
    print({i}, X_train[i].isna().sum())
```

```python
col = X_test.columns
for i in col:
    print({i}, X_test[i].isna().sum())
```

```python
# 이상치 처리 (카테고리: 마스킹, 분리 등)
```

```python
# 이상치 처리 (수치: 비율화 등)
percent_col = ['Top10perc','Top25perc','PhD','Terminal','S.F.Ratio','perc.alumni', 'Grad.Rate']
X_train[percent_col] = X_train[percent_col]/100
X_test[percent_col] = X_test[percent_col]/100
```

```python
# 상관관계 분석하여 칼럼 축소
X_train.corr()
```

```python
# Accept - Apps
# Enroll - Accept
# F.Undergrad - Enroll
# Top25perc - Top10perc
# Terminal - phD

# => Accept, Enroll, top10perc, Terminal

drop_col = ['Accept','Enroll','Top10perc','Terminal']

X_train.drop(columns=drop_col, inplace=True)
X_test.drop(columns=drop_col, inplace=True)

print(X_train.shape)
print(X_test.shape)
```

```python
# 카테고리 데이터 인코딩
X_test.head() # 카테고리 데이터 없음
```

```python
# 수치형 데이터 정규화
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_std = pd.DataFrame(scaler.fit_transform(X_train))
X_test_std = pd.DataFrame(scaler.fit_transform(X_test))
```

```python
# 데이터 분할 (train으로 train과 val)
from sklearn.model_selection import train_test_split

train_x, val_x, train_y, val_y = train_test_split(X_train, y_train, test_size = 0.2, random_state = 2025)

print(train_x.shape)
print(train_y.shape)
print(val_x.shape)
print(val_y.shape)
```

```python
# 모델 생성 및 학습
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(train_x, train_y)
```

```python
# 모델 평가 및 선정
from sklearn.metrics import roc_curve, auc
y_score = model.predict_proba(val_x)[:,1]
fpr, tpr, thresholds = roc_curve(val_y, y_score, pos_label='Yes')
AUC = auc(fpr, tpr)

print(AUC)
```

```python
# 모델로 최종 예측 (test 사용)
y_pred = pd.DataFrame(model.predict(X_test_std))
```

```python
# id와 예측 결과 합쳐서 csv화
result = pd.concat([result_id, y_pred], axis=1).rename(columns={0: 'prob_Private'})
result.to_csv(r'C:\BigData\test\연습문제\작업형2\result\result_2.csv')
```
